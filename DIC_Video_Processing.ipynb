```python
import cv2
import os

# 打开视频
vid = cv2.VideoCapture("F:\Vectran_UniaxialTension\Warp_4000N_Creep\DIC_Image\Warp_4000N_Creep_1\DSC_0029.MOV")

# 获取视频的帧率
fps = vid.get(cv2.CAP_PROP_FPS)

# 设置抽帧间隔（秒）
interval_sec = 60   # 每隔 2 秒抽一帧

# 计算多少帧对应 interval_sec 秒
frame_interval = int(fps * interval_sec)

# 创建保存目录
save_dir = "F:\Vectran_UniaxialTension\Warp_4000N_Creep\DIC_Image\Warp_4000N_Creep_1\part3"
os.makedirs(save_dir, exist_ok=True)

count = 0      # 总帧计数
save_idx = 1   # 保存图片编号（1, 2, 3...）

while True:
    ret, frame = vid.read()
    if not ret:
        break
    if count % frame_interval == 0:  # 每隔固定时间保存一张
        cv2.imwrite(os.path.join(save_dir, f"{save_idx}.jpg"), frame)
        save_idx += 1
    count += 1

vid.release()
```
