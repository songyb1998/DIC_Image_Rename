```python
import os
import shutil

# 将文件夹中的照片进行重新命名

def ReFileName(dirPath, outPath):
    # Check if output path exists, if not, create it
    if not os.path.exists(outPath):
        os.makedirs(outPath)

    # Get files and sort them by the numeric value in the filename
    files = [file for file in os.listdir(dirPath) if file.startswith("DSC_") and os.path.isfile(os.path.join(dirPath, file))]
    
    # 自定义排序：提取文件名中的数字并按数值排序
    files.sort(key=lambda x: int(x.split("_")[1].split(".")[0]))  

    # Initialize counter for new file names
    counter = 1

    # Iterate over sorted files
    for file in files:
        # Extract file extension
        file_extension = os.path.splitext(file)[1]
        
        # Construct the new file name
        newfile = f"Image_{counter}{file_extension}"
        
        # Copy and rename file
        shutil.copy(os.path.join(dirPath, file), os.path.join(outPath, newfile))
        
        # Increment counter
        counter += 1

    print("*** Done! ***")

# Specify your own paths here
dirPath = r"C:\Users\songy\Desktop\75-10mm"  # Input folder path
outPath = r"C:\Users\songy\Desktop\75-10mm-new"  # Output folder path
ReFileName(dirPath, outPath)
```

```python
import os

# 递归遍历文件夹，减少图片数量
def reduce_images(folder, max_images=100):
    # 获取所有图片
    files = [f for f in os.listdir(folder) if f.lower().startswith("image_") and os.path.isfile(os.path.join(folder, f))]
    
    if not files:
        return
    
    # 按编号排序
    files.sort(key=lambda x: int(os.path.splitext(x)[0].split("_")[1]))

    while len(files) > max_images:
        keep = [files[0], files[-1]]  # 保留首尾
        keep += files[1:-1:2]         # 中间隔一取一
        keep = sorted(set(keep), key=lambda x: int(os.path.splitext(x)[0].split("_")[1]))
        
        # 删除不在 keep 中的文件
        for f in files:
            if f not in keep:
                os.remove(os.path.join(folder, f))
        
        files = keep
        print(f"  └ Reduced to {len(files)} images")

    print(f"  └ Final kept: {len(files)} images")


def process(root_dir):
    # 获取所有子文件夹
    subfolders = [sub for sub in os.listdir(root_dir) if os.path.isdir(os.path.join(root_dir, sub))]
    total = len(subfolders)

    # 遍历所有子文件夹，带进度显示
    for idx, sub in enumerate(subfolders, start=1):
        sub_path = os.path.join(root_dir, sub)
        print(f"\nProcessing {idx}/{total}: {sub}")
        reduce_images(sub_path, max_images=100)

    print("\n*** All processing completed! ***")


if __name__ == "__main__":
    root_dir = r"F:\Vectran_UniaxialTension\Warp_4000N_Creep\DIC_Image"
    process(root_dir)
```

```python
# 将某个文件夹中的照片旋转一定的角度

from PIL import Image
import os

# 设置文件夹路径（原始图片所在文件夹）
input_folder = r"F:\Vectran_UniaxialTension\Warp_4000N_Creep\DIC_Image\Warp_4000N_Creep_30"
# 设置输出文件夹（旋转后的图片存放位置）
output_folder = r"F:\Vectran_UniaxialTension\Warp_4000N_Creep\DIC_Image\Warp_4000N_Creep_30_rotated"
# 设置旋转角度（顺时针为负，逆时针为正）
rotate_angle = 0.1

# 确保输出文件夹存在
os.makedirs(output_folder, exist_ok=True)

# 遍历文件夹中的所有 JPG 图片
for filename in os.listdir(input_folder):
    if filename.lower().endswith(".jpg"):  # 确保是 JPG 文件
        img_path = os.path.join(input_folder, filename)
        img = Image.open(img_path)  # 打开图片
        rotated_img = img.rotate(rotate_angle, expand=True)  # 旋转并自动调整大小
        save_path = os.path.join(output_folder, filename)  # 生成保存路径
        rotated_img.save(save_path)  # 保存图片
        print(f"已旋转 {filename} 并保存到 {save_path}")

print("所有 JPG 图片处理完成！")
```

```python
# 将某个文件夹中的照片保留ROI，并将其余部分涂黑

from PIL import Image
import os

# 设置文件夹路径（原始图片所在文件夹）
input_folder = r"F:\Vectran_BiaxialTension\BTS(3-1)\DIC_Image\BTS(3-1)_1"
# 设置输出文件夹（处理后的图片存放位置）
output_folder = r"F:\Vectran_BiaxialTension\BTS(3-1)\DIC_Image\BTS(3-1)_1_Blacked"
# 定义要保留的区域 (left, upper, right, lower)
region_to_keep = (2432, 1377, 3634, 2566)  # 根据需求调整

# 确保输出文件夹存在
os.makedirs(output_folder, exist_ok=True)

# 遍历文件夹中的所有 JPG 图片
for filename in os.listdir(input_folder):
    if filename.lower().endswith(".jpg"):  # 确保是 JPG 文件
        img_path = os.path.join(input_folder, filename)
        img = Image.open(img_path)  # 打开图片

        # 创建一个与原始图片大小相同的纯黑背景
        black_background = Image.new('RGB', img.size, (0, 0, 0))

        # 裁剪出要保留的区域
        cropped_region = img.crop(region_to_keep)

        # 将裁剪的区域粘贴到纯黑背景上
        black_background.paste(cropped_region, region_to_keep[:2])

        # 生成保存路径
        save_path = os.path.join(output_folder, filename)

        # 保存处理后的图片
        black_background.save(save_path)
        print(f"已处理 {filename} 并保存到 {save_path}")

print("所有 JPG 图片处理完成！")
```

```python
import os

# 转换图片后缀统一为jpg
# 要处理的根目录
root_dir = r"F:\Vectran_UniaxialTension\Warp_4000N_Creep\DIC_Image"

# 支持的图片扩展名（都改为小写 jpg）
valid_exts = [".JPG", ".JPEG", ".PNG", ".BMP", ".TIF", ".TIFF"]

for subdir, dirs, files in os.walk(root_dir):
    for file in files:
        file_path = os.path.join(subdir, file)
        name, ext = os.path.splitext(file)
        if ext.upper() in valid_exts:
            new_name = name + ".jpg"
            new_path = os.path.join(subdir, new_name)
            # 如果新名字和旧名字不同才改名，避免重复改名报错
            if new_path != file_path:
                os.rename(file_path, new_path)
                print(f"Renamed: {file_path} -> {new_path}")

print("All file extensions have been converted to .jpg")
```
