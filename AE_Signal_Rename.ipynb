```python
# This script renames folders and files in a specified directory according to a specific naming convention.

import os
import re

# 设置主文件夹路径
main_folder_path = r"F:\Vectran_BiaxialTension\BTS(3-1)\AE_Signal"

# 后缀生成器（避免使用 _1, _2 等，使用字母后缀）
def generate_unique_suffix(existing_files, base_name, extension):
    suffix_counter = 0
    suffix_chars = "abcdefghijklmnopqrstuvwxyz"
    while True:
        suffix = f"-{suffix_chars[suffix_counter]}"
        new_filename = f"{base_name}{suffix}{extension}"
        if new_filename not in existing_files:
            return suffix
        suffix_counter += 1
        if suffix_counter >= len(suffix_chars):
            suffix_counter = 0  # 循环使用字母

# 遍历主文件夹中的所有子文件夹
for folder_name in os.listdir(main_folder_path):
    folder_path = os.path.join(main_folder_path, folder_name)
    # 确保是文件夹且符合命名规则（例如：BTS(1-2)_1 到 BTS(1-2)_35）
    if os.path.isdir(folder_path) and folder_name.startswith("BTS(3-1)_"):
        match = re.match(r"BTS\(3-1\)_(\d+)", folder_name)
        if match:
            folder_number = int(match.group(1))
            new_folder_name = f"BTS(3-1)_{folder_number}"  # 生成新的文件夹名称
            new_folder_path = os.path.join(main_folder_path, new_folder_name)
            
            try:
                os.rename(folder_path, new_folder_path)
                print(f"重命名文件夹: {folder_name} -> {new_folder_name}")
                
                for filename in os.listdir(new_folder_path):
                    file_path = os.path.join(new_folder_path, filename)
                    if os.path.isfile(file_path):
                        file_extension = os.path.splitext(filename)[1]
                        base_name = new_folder_name
                        # 检查文件名是否已经和子文件夹名称相同（忽略后缀）
                        if os.path.splitext(filename)[0] == base_name:
                            print(f"文件名已满足条件，跳过: {filename}")
                            continue  # 跳过该文件的修改
                        
                        new_filename = f"{base_name}{file_extension}"
                        existing_files = set(os.listdir(new_folder_path))
                        
                        if new_filename in existing_files:
                            suffix = generate_unique_suffix(existing_files, base_name, file_extension)
                            new_filename = f"{base_name}{suffix}{file_extension}"
                        
                        new_file_path = os.path.join(new_folder_path, new_filename)
                        os.rename(file_path, new_file_path)
                        print(f"修改文件名: {filename} -> {new_filename}")
                        
            except ValueError as e:
                print(f"文件夹名格式不正确，跳过: {folder_name}。错误信息: {e}")
```

```python
import os
import re

# 主文件夹路径
main_folder_path = r"F:\Vectran_UniaxialTension\Warp_4000N_Creep\AE_Signal"

# 生成唯一后缀（-a, -b, -c ...）
def generate_suffix(counter):
    suffix_chars = "abcdefghijklmnopqrstuvwxyz"
    return f"-{suffix_chars[counter % len(suffix_chars)]}"

# 遍历所有子文件夹
for folder_name in os.listdir(main_folder_path):
    folder_path = os.path.join(main_folder_path, folder_name)

    # 只处理符合命名规则的文件夹 Warp_4000N_Creep_x
    if os.path.isdir(folder_path) and folder_name.startswith("Warp_4000N_Creep_"):
        print(f"处理文件夹: {folder_name}")

        # 收集所有文件并按扩展名分组
        files_by_ext = {}
        for filename in os.listdir(folder_path):
            file_path = os.path.join(folder_path, filename)
            if os.path.isfile(file_path):
                # 文件扩展名：取第一个点之后的所有
                parts = filename.split(".", 1)
                if len(parts) < 2:
                    continue  # 跳过没有扩展名的文件
                extension = "." + parts[1]
                files_by_ext.setdefault(extension, []).append(filename)

        # 遍历每个扩展名组
        for ext, files in files_by_ext.items():
            # 按文件名长度排序（短的优先）
            files_sorted = sorted(files, key=lambda x: len(x))

            for idx, filename in enumerate(files_sorted):
                file_path = os.path.join(folder_path, filename)

                # 基础新名字 = 文件夹名 + 扩展名
                if idx == 0:
                    new_filename = f"{folder_name}{ext}"
                else:
                    new_filename = f"{folder_name}{generate_suffix(idx-1)}{ext}"

                new_file_path = os.path.join(folder_path, new_filename)

                # 避免重名覆盖
                if os.path.exists(new_file_path):
                    print(f"⚠️ 文件已存在，跳过: {new_filename}")
                    continue

                os.rename(file_path, new_file_path)
                print(f"重命名: {filename} -> {new_filename}")
```
